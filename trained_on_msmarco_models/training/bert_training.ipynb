```python
!pip install import-ipynb
%cd '/content/drive/MyDrive/QA_training/generative_QA'
import import_ipynb
from support_functions import *
```

```python
tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')
```

```python
number = 4
```

```python
train_csv_path = dataset_paths[number]
print('Dataset path:', train_csv_path)

if number in [0, 2]:
    text_num = 1  
elif number >= 3: 
    text_num = 3
else:
    text_num = 2
print('Text num =', text_num)

train_dataset = prepare_data(train_csv_path, tokenizer,  text_num=text_num, parts_num = 2, part = 1)
```

```python
dir = model_paths[number+1]

print('Importing model path:', model_paths[number])
print('New model path:', dir)

#model = EncoderDecoderModel.from_encoder_decoder_pretrained("bert-base-uncased", "bert-base-uncased", bos_token_id=101, eos_token_id=102, pad_token_id=0)
model = EncoderDecoderModel.from_pretrained(model_paths[number])
```

```python
training_args = TrainingArguments(
    output_dir=dir,          # output directory
    num_train_epochs=2,              # total number of training epochs
    per_device_train_batch_size=8,  # batch size per device during training
    per_device_eval_batch_size=8,   # batch size for evaluation
    logging_dir=dir+'/logs',            # directory for storing logs
    logging_steps=1000,
    save_steps = 1000,
    prediction_loss_only=False,
    overwrite_output_dir = True,
    learning_rate = 1e-4
)


trainer = Trainer(
    model=model,                         # the instantiated 🤗 Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,         # training dataset
    #eval_dataset=val_dataset,            # evaluation dataset

)
```

```python
print('Dataset path:', train_csv_path)
print('New model path:', dir)
trainer.train()
#trainer.train(dir+'/checkpoint-8000')
trainer.save_model(dir)
```
